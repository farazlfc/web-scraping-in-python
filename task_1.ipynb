```python
def string_mod(string):
    string = list(string)
    for i,j in enumerate(string):
        if(j == ","):
            string[i] = " "

    string = ("".join(string))
    
    return " ".join(string.split())
    
    

            
string_mod("faraz , jamal       zaidi")
```

```python
Inventors = []
for i in range(1,11):
    soup = BeautifulSoup(open(r"C:\Users\Farz Jamal\Desktop\us\us_%d.html"%i) , "html.parser")
    for element in soup.find_all('td'):
        try:
            if (element['align'],element['width']) == ('left','90%'):
                names = []
                for i in (element.find_all('b')[::2]):
                    text = i.get_text().split(";")
                    text = " ".join([txt.strip() for txt in text][::-1])
                    
                    names.append(string_mod(text))
                    
                #print element
                break
        except:
            pass
    Inventors.append(names)
    print(names)
print(Inventors)
```
